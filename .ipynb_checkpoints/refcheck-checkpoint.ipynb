## Etape 4 - Retrieve

Input : source.json -> Fichier JSON avec les informations des références bibliographiques extraites du document original

Output : trusted.json -> Fichier JSON avec les métadonnées extraits depuis des bases de données internationales 

Méthode :
1. Recherche des identifiants uniques (PIDs) : DOI et PMID
2. Pour les réfs avec PID : utiliser une API pour obtenir les métadonnées complets
   * DOI -> Crossref
   * PMID -> PubMed
4. Pour les références sans PID : construire une requête basée sur le titre + auteur + année de publicaiton
   * Requête booléenne -> OpenAlex
5. Extraire les informations essentielles et produire le fichier trusted.json
   * CrossRef
       * title : $.message.title
       * author : $.message.author
       * journal : $.message.container-title
       * year : message.issued.date-parts
       * pages : $.message.page
       * volume : $.message.volume
       * DOI : $.message.DOI
